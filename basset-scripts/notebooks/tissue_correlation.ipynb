```python
%pylab inline
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
import pandas as pd
import numpy as np
import matplotlib.patches as mpatches
```

```python
# Read in meta data
meta = pd.read_csv("../celltype_metadata.csv")

colors = {
    "development": "red",
    "cancer": "yellow",
    "blood": "purple",
    "endothelial": "orange",
    "nervous": "green",
    "fibroblast": "blue",
    "skin": "blue",
    "melanocyte": "blue",
    "keratinocyte": "blue",
    "heart": "cyan",
    "liver": "cyan",
    "pancreas": "cyan",
    "lung": "cyan"
}
meta["color"] = meta.category.apply(lambda x: colors.get(x, "gray"))
meta.index = meta.tissue
meta.head()
```

```python
# Read in corr matrix
corrmat = pd.read_csv("../tmpdata/tissue_corr.tab", sep="\t")
corrmat = corrmat.drop("Unnamed: 0", 1)
corrmat.index = corrmat.columns

# Which columns to keep?
tokeep = list(meta[(meta["category"].apply(lambda x: x not in ["cancer","treatment","disease"])) &
                   (meta["color"]!= "gray")].tissue)
corrmat = corrmat.loc[tokeep,tokeep]
```

```python
# Cluster and plot
D = corrmat.as_matrix()
Y = sch.linkage(D, method='centroid')
Z = sch.dendrogram(Y, orientation='right', no_plot=True)
index = Z['leaves']
D = D[index,:]
D = D[:,index]
tissues = np.array(corrmat.columns)[index]

fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(111)
im = ax.matshow(D, aspect='auto', origin='lower')
ax.set_xticks([])
ax.set_yticks(range(D.shape[0]))
ax.set_yticklabels(tissues, size=8);

# Add color bar
colors = list(meta.loc[tissues,:].color)
for i in range(D.shape[0]):
    ax.add_patch(mpatches.Rectangle((i-0.5, D.shape[0]), 1, 1, color=colors[i]))
ax.set_ylim(bottom=0, top=D.shape[0]+1);
```
